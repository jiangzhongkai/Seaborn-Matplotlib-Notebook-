###  利用Seaborn进行数据可视化 

#### 首先是导入相应的库

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
sns.set()
```

#### 载入数据

```python
data=pd.read_csv("train.csv",encoding="gb18030")
```

```python
data.head(5)
```

##### 重命名某列名

```python
data=data.rename(columns={'收率':'target'})
data.columns
```

#### 绘制直方图
##### 利用Distplot来对上述数据的某列特征进行直方图展示

```python
sns.distplot(data['target'],bins=32,color='purple')
```

######   上述图片显示进行简单的介绍：
sns.distplot-表示将启动distplot的初始创建

data['target]表示从数据集中取出相应的列，并显示出来

##### 绘制联合分布图
主要就是采用将两个变量并在一起创建直方图和散点图

```python
sns.jointplot(x='B12',y='target',data=data)
```

上述图可以看出，变量B12和target之间的相关性

##### 绘制矩阵图
矩阵图是绘制了变量之间的成对关系，语法是sns.pairplot("DataFrame"),并且支持用“hue”来未变量绘图着色

```python
sns.pairplot(data)
```

##### 绘制条形图
条形图主要是绘制分类列和数字列，在可视化中创建了条形，语法：sns.barplot(x,y,data)

例如：我们用条形图来显示B14的分布与target之间的关系

```python
sns.barplot(x='B14',y='target',data=data)
```

##### 绘制箱形图
箱形图是给定数据集的5点汇总统计的直观表示，5个数字分别是：
最小值；
1/4值；
2/4值(中位数)；
3/4值；
最大值；

语法：sns.boxplot(x,y,data)

```python
sns.boxplot(x='B14',y='target',data=data)
```

#####  线性回归的拟合线
语法：sns.lmplot(x,y,data)

遵循最小二乘法，线代表最佳拟合线

```python
sns.lmplot(x='B14',y='target',data=data)
```
